```python
%config IPCompleter.greedy=True

import os
import pandas as pd

data = pd.DataFrame(columns=['File Path', 'Row Count'])
out_path = "E:/downsample.invitrogen.new/fastq/"

file_list =  [file for file in os.listdir(out_path) if file.endswith('.fastq')]

print(file_list)
```

```python
for file in file_list:
    file_path = os.path.join(out_path, file) # Get the full file path
    with open(file_path, 'r') as f:
        lines = f.readlines() # Read all lines in the file
        row_count = len(lines) # Count the number of lines in the file
    # Append the file path and row count to the DataFrame
    data = data.append({'File Path': file_path, 'Row Count': row_count}, ignore_index=True)

# Display the resulting DataFrame
print(data)
```

```python
print(data.iloc[:,1].max())
values = [5000, 10000, 20000, 50000, 75000, 100000, 125000, 150000, 175000, 200000, 225000, 250000, 500000, 750000, 1000000, 1500000, 2000000, 2500000]
result = []
for index, row in data.iterrows():
    col2_value = row['Row Count']
    largest_value = None
    for value in values:
        if value < col2_value and value % 2 == 0:
            largest_value = value
    result.append(largest_value)

data['greater_than'] = result

print(data)
```

```python
import random

names = data['greater_than'].unique().astype(str)
print(names)
outputfolder = str("D:/data/Totiteszt/kaiju/result.v2/")
for folder in names:
    folder_path = os.path.join(outputfolder, folder)
    os.makedirs(folder_path, exist_ok=True)

for row in range(0, len(data)):
    find = data.iloc[row, 2]
    path = data.iloc[row, 0]
    filename = os.path.basename(data.iloc[row, 0]).split('.')[0]
    print(filename)
    print(path)
    print(find)
    commands = [value for value in values if value <= find]
    print(commands)
    for command in commands:
        outputfilename = filename + "_" + str(command) + "_downsample.out"
        output = outputfolder + str(find) + "/" + outputfilename
        print(output)
        with open(path, 'r') as file:
            lines = file.readlines()
            print(command)
            randomreads = random.sample(lines, k=command)
            print(len(randomreads))
        with open(output, 'w') as file:
            file.writelines(randomreads)
```
